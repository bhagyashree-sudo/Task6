```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset=pd.read_csv("/content/car_data.csv")

x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values

from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(x,y)

new_car=[[1600,4,16]]
predicted_price=rf.predict(new_car)
print("Predicted car price:",predicted_price)
engine=dataset.iloc[:,0].values.reshape(-1,1)
x_grid=np.arange(engine.min(),engine.max(),10).reshape(-1,1)
prices=[rf.predict([[e[0],4,15]])for e in x_grid]
plt.scatter(dataset.iloc[:,0],y)
plt.plot(x_grid,prices)
plt.title("Random Forest Regression - Car dataset")
plt.xlabel("Engine Size(cc)")
plt.ylabel("Car Price(lakhs)")
plt.show()
```
